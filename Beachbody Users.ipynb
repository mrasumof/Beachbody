```python
__author__ = 'martinrasumoff'

import requests
import cnfg
from pymongo import MongoClient
from requests_oauthlib import OAuth1


config = cnfg.load(".twitter_config")

oauth = OAuth1(config["consumer_key"],
               config["consumer_secret"],
               config["access_token"],
               config["access_token_secret"])

client = MongoClient("52.25.140.201",27017)

db = client['Beachbody']
coll_tweets = db['tweets']
```

```python
#import pprint
# { "user_location": { $exists: false } } )

from random import randint
from time import sleep

count = 0
for each in coll_tweets.find({"user_location": {"$exists":"False"}}):
    try: 
        a = each['user_location']
    except KeyError:
        pass

    tweet_usr = each['user_screen_name']

    string = "https://api.twitter.com/1.1/users/search.json?q="+tweet_usr
    #parameters = {"q": tweet_usr}

    response = requests.get(string, auth=oauth)
    times = 0

    while (response.status_code != 200):
        print "Oops...Trying again..."+str(times)+" times - "+str(response)
        response = requests.get(string, auth=oauth)
        secs = randint(3,15)
        print "Waiting "+str(secs)+" seconds"
        sleep(secs)
        times += 1

    tweets = response.json()

    try:
        usr_location = tweets[0]['location']
    except IndexError:
        pass
    except KeyError:
        pass

    usr_description = tweets[0]['description']
    usr_follow_count = tweets[0]['followers_count']
    usr_time_zone = tweets[0]['time_zone']

#    print usr_location
#    print usr_description
#    print usr_follow_count
#    print usr_time_zone
#    print each['status_id']
#    print "-----------------------"

    get_id = each['status_id']
#    print get_id
    coll_tweets.update({"status_id" : get_id}, {"$set":{"user_location": usr_location,
                                                        "usr_description": usr_description,
                                                        "usr_follow_count": usr_follow_count,
                                                        "usr_time_zone": usr_time_zone}})
    count += 1
    print "Processed: "+str(count)
```
